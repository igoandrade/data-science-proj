# Teste de significância

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
from pylab import rcParams
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
```

```python
def plot_distribution(inp):
    plt.figure()
    ax = sns.displot(inp, kde=True)
    plt.axvline(np.mean(inp), color='k', linestyle='dashed', linewidth=2, label=f'Mean: {inp.mean():.6f}')
    plt.legend()
    return ax
```

```python
pre = df['2018-01-01':'2019-12-31']['Log Adj Return'].dropna().to_numpy()
pos = df['2020-01-01':]['Log Adj Return'].dropna().to_numpy()
```

```python
ax = plot_distribution(pre)
```

```python
plot_distribution(pos)
```

```python
def compare_2_groups(arr_1, arr_2, alpha):
    stat, p = ttest_ind(arr_1, arr_2)
    print(f'Statistics = {stat:.3f}, p = {p:.3f}')
    if p > alpha:
        print('Same distributions (fail to reject H0)')
    else:
        print('Different distributions (reject H0)')
```

```python
compare_2_groups(pre, pos, 0.05)
```

```python
F, p = f_oneway(pre, pos)
```

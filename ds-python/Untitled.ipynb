```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.read_csv('convertcsv.csv')
```

```python
df
```

```python
df.columns
```

```python
corr = df.corr()
#corr.iloc[0,1]
corr
```

```python
ax = df.plot.scatter(x = 'idade', y = 'tempo (minutos)')
ax.set_xlabel('Idade')
ax.set_ylabel('Tempo de Conclusão (min.)')
plt.annotate(f'corr = {corr.iloc[0,1]:.2f}', (22.1, 48.6), fontsize=11)
plt.show()
```

```python
ax = df.plot.scatter(x = 'tempo de treinamento (anos)', y = 'tempo (minutos)')
ax.set_xlabel('Tempo de treinamento (anos)')
ax.set_ylabel('Tempo de Conclusão (min.)')
plt.annotate(f'corr = {corr.iloc[0,2]:.2f}', (12.2, 48.6), fontsize=11)
plt.show()
```
